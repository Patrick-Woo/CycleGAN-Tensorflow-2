```python
from mmdet.apis import init_detector, inference_detector, show_result_pyplot
import mmcv
```

```python
# ssd_regnet
config_file = '../configs/edgeailite/ssd/ssd_regnet_fpn_bgr_lite.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
# url: https://download.openmmlab.com/mmdetection/v2.0/faster_rcnn/faster_rcnn_r50_fpn_1x_coco/faster_rcnn_r50_fpn_1x_coco_20200130-047c8118.pth
checkpoint_file = '../checkpoints/ssd_regnetx-800mf_fpn_bgr_lite_512x512_20200919_checkpoint.pth'
```

```python
# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')
#model = init_detector(config_file, checkpoint_file, device='cuda:0', cfg_options=dict(test_cfg=dict(score_thr=0.3)))
```

```python
# test a single image
img = 'demo.jpg'
result = inference_detector(model, img)
```

```python
# show the results
show_result_pyplot(model, img, result)
```
